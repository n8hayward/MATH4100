## Exercise 1: Reading and Writing Data

The file [grades.csv](grades.csv) is a file with student names and letter grades:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Read the file into an array. Add a GPA to the student's row (A=4,B=3,C=2,D=1). 

Hint: the function [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) removes trailing whitespace from a string.

Write that file into a new file `grades_gpa.csv`

````python
gpas = {"A":4, "B":3, "C":2, "D":1}
````

````python
#import csv livrary
import csv 

#dictionary for gpas
gpas = {"A":4, "B":3, "C":2, "D":1}

# initialize the top-level array
gpa_data = []

# open the semi-colon-delimited file
with open('grades.csv', 'r') as csvfile:
    filereader = csv.reader(csvfile, delimiter=';', quotechar="'")
    for row in filereader:
            name, grade = row #separate row into name and grade
            gpa = gpas.get(grade.strip())  # Get gpa from gpas dictionary with key being grad.strip() to ensure there is no leading white space and default 
            gpa_data.append([name.strip(), grade.strip(), gpa]) #append gpa_data array with name, grad, gpa to create new array with gpa data

print(gpa_data)

# write to the new file
with open('grades_gpa.csv', 'w', newline = '') as csvfile:
    filewriter = csv.writer(csvfile, delimiter = ',', quotechar="'") #writes csv with comma delimited format as above
    for rows in gpa_data:
        filewriter.writerow(rows) #writes each row to grades_gpa.csv file

print("File 'grades_gpa.csv' created successfully!") #Show that code ran successfully
````
